```python
from prettytable import PrettyTable

# Input student details
student_name = input("Enter Student Name: ")
roll_number = input("Enter Roll Number: ")

# Subjects and marks input
subjects = ["Math", "Science", "English", "History", "Computer"]
marks = []
total_marks = 500  # Assuming each subject is out of 100

print("\nEnter marks out of 100 for each subject:")
for subject in subjects:
    mark = int(input(f"Marks in {subject}: "))
    marks.append(mark)

# Calculate total and percentage
obtained_marks = sum(marks)
percentage = (obtained_marks / total_marks) * 100

# Create the marksheet table
table = PrettyTable()
table.field_names = ["Subject", "Marks Obtained", "Total Marks"]
for i in range(len(subjects)):
    table.add_row([subjects[i], marks[i], 100])

# Display student details and marksheet
print("\nMarksheet")
print(f"Student Name: {student_name}")
print(f"Roll Number: {roll_number}\n")
print(table)

# Display total marks and percentage
print(f"Total Marks: {obtained_marks}/{total_marks}")
print(f"Percentage: {percentage:.2f}%")

from prettytable import PrettyTable

# Header for the marksheet
def display_heading():
    print("=" * 50)
    print("   Sindh Board of Technical Education Karachi   ")
    print("=" * 50)

# Input student details
student_name = input("Enter Student Name: ")
roll_number = input("Enter Roll Number: ")

# Subjects and marks input
subjects = ["Math", "Science", "English", "History", "Computer"]
marks = []
total_marks = 500  # Assuming each subject is out of 100

print("\nEnter marks out of 100 for each subject:")
for subject in subjects:
    mark = int(input(f"Marks in {subject}: "))
    marks.append(mark)

# Calculate total and percentage
obtained_marks = sum(marks)
percentage = (obtained_marks / total_marks) * 100

# Create the marksheet table
table = PrettyTable()
table.field_names = ["Subject", "Marks Obtained", "Total Marks"]
for i in range(len(subjects)):
    table.add_row([subjects[i], marks[i], 100])

# Display the full marksheet
print("\n" + "=" * 50)
display_heading()  # Display the heading
print(f"Student Name: {student_name}")
print(f"Roll Number: {roll_number}")
print("=" * 50)
print(table)
print("=" * 50)
print(f"Total Marks: {obtained_marks}/{total_marks}")
print(f"Percentage: {percentage:.2f}%")
print("=" * 50)
```
